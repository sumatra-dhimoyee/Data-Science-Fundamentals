c) In your own words, briefly explain:

- underfitting
- overfitting

and what signs to look out for for each.

When the model cannot identify the underlying patterns in the data, it is said to be underfitting. Poor performance on the training and validation/test datasets is a sign of underfitting. High bias and low variance in the model could result in simplistic predictions on test sets that don't adequately reflect the issue's complexity.

When a model learns too much from the training set, it may become overfit and begin to interpret noise and random variations as significant patterns. High performance on the training set and noticeably poorer performance on the validation/test set are indicators of overfitting. Low bias but high variance in the model could result in poorly generalizable and highly complicated predictions.

Outliers are points that lie far from the rest of the data. They are not necessarily invalid points however. Imagine sampling from a Normal Distribution with mean 10 and variance 1. These outliers can inhibit our ability to learn general patterns in the data since they are not representative of likely outcomes.
